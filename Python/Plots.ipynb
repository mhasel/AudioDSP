```python
import matplotlib.pyplot as plt
import numpy as np
import dsp_helpers as dsp
from analysis_helpers import ResultsPath

plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = (9, 6)
destination = ResultsPath()
x = dsp.sine_wave(1, 1000, 48000, 1)
t = dsp.time(x, 48000)
```

```python
yAxisLabel = 'amplitude, normalized'
xAxisLabel = 'time in milliseconds'
yOverdrive = dsp.normalize(dsp.overdrive(x))

plt.plot(t[0:500], x[0:500])
plt.title('Test signal')
plt.xlabel(xAxisLabel)
plt.ylabel(yAxisLabel)
plt.grid()
#plt.savefig(destination + 'test_signal.png')
plt.show()


plt.plot(t[0:500], yOverdrive[0:500])
plt.title('Overdrive')
plt.xlabel(xAxisLabel)
plt.ylabel(yAxisLabel)
plt.grid()
#plt.savefig(destination + 'test_overdrive.png')
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (9,12)
fig, (ax1, ax2) = plt.subplots(2, sharey = True)
fig.tight_layout(pad=3.0)
ax1.plot(t[0:500], x[0:500])
ax2.plot(t[0:500], yOverdrive[0:500])
ax1.grid()
ax2.grid()
ax1.set_title('Test signal')
ax2.set_title('Overdrive')
ax1.set_ylabel(yAxisLabel)
ax2.set_ylabel(yAxisLabel)
plt.xlabel(xAxisLabel)
plt.savefig(destination + 'overdrive_test')
plt.show()
```

```python
yFuzz = dsp.normalize(dsp.fuzz(x, gain=11, mix = 0.9))

fig, (ax1, ax2) = plt.subplots(2, sharey = True)
fig.tight_layout(pad=3.0)
ax1.plot(t[0:500], x[0:500])
ax2.plot(t[0:500], yFuzz[0:500])
ax1.grid()
ax2.grid()
ax1.set_title('Test signal')
ax2.set_title('Fuzz')
ax1.set_ylabel(yAxisLabel)
ax2.set_ylabel(yAxisLabel)
plt.xlabel(xAxisLabel)
plt.savefig(destination + 'fuzz_test')
plt.show()
```

```python
yRingModSin = dsp.normalize(dsp.ring_modulator(x, 0, rate=.9, blend=.9))
yRingModTriangle = dsp.normalize(dsp.ring_modulator(x, 1, rate=.9, blend=.9))
yRingModSquare = dsp.normalize(dsp.ring_modulator(x, 2, rate=.9, blend=.9))

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, sharey = True)
fig.tight_layout(pad=3.0)
ax1.plot(t[0:3500], x[0:3500])
ax2.plot(t[0:3500], yRingModSin[0:3500])
ax3.plot(t[0:3500], yRingModTriangle[0:3500])
ax4.plot(t[0:3500], yRingModSquare[0:3500])
ax1.grid()
ax2.grid()
ax3.grid()
ax4.grid()
ax1.set_title('Test signal')
ax2.set_title('Ring modulator: Sine wave')
ax3.set_title('Ring modulator: Triangle wave')
ax4.set_title('Ring modulator: Square wave')
ax1.set_ylabel(yAxisLabel)
ax2.set_ylabel(yAxisLabel)
ax3.set_ylabel(yAxisLabel)
ax4.set_ylabel(yAxisLabel)
plt.xlabel(xAxisLabel)
plt.savefig(destination + 'ringmod_test')
plt.show()
```

```python
yTremolo = dsp.normalize(dsp.tremolo(x, rate=.9, depth=.9))

fig, (ax1, ax2) = plt.subplots(2, sharey = True)
fig.tight_layout(pad=3.0)
ax1.plot(t[0:4500], x[0:4500])
ax2.plot(t[0:4500], yTremolo[0:4500])
ax1.grid()
ax2.grid()
ax1.set_title('Test signal')
ax2.set_title('Tremolo')
ax1.set_ylabel(yAxisLabel)
ax2.set_ylabel(yAxisLabel)
plt.xlabel(xAxisLabel)
plt.savefig(destination + 'tremolo_test')
plt.show()
```
